```python
import pandas as pd

# Function to convert height to inches
def convert_height(height):
    try:
        # Check if height is a string and contains '-' or "'"
        if isinstance(height, str) and ('-' in height or "'" in height):
            height = height.replace("'", "-").strip()  # Normalize to "6-7"
            feet, inches = height.split('-')
            return int(feet) * 12 + int(inches)
        return 0  # Return 0 for invalid or non-string values
    except:
        return 0  # Return 0 if conversion fails

# Apply conversion to height column
PD['height_in_inches'] = PD['height'].apply(convert_height)

# Find the player with the maximum height
tallest = PD.nlargest(1, 'height_in_inches')[['name', 'height']]

# Print result
print(f"Tallest player: {tallest['name'].iloc[0]}, Height: {tallest['height'].iloc[0]}")
```
